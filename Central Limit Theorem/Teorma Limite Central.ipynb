```python
# vamos importar algumas bibliotecas que serão utilizadas no exercício!
import numpy as np
import scipy.stats as sct
import seaborn as sns
import random
```

```python
# lista de números que vamos avaliar

lista = [3, 5, 6, 6, 6, 6, 6, 7, 10, 10, 13, 13, 14, 16, 17, 17, 18, 20, 20, 20]
```

```python
randomlist = []
for i in range(1,1001):
    n=random.randint(1,30)
    randomlist.append(n)
print(randomlist)

len(randomlist)
```

```python
# Vamos gerar todas as distribuições considerando que para todos a localização é 10 da média e a distância da média (largura) é 2

# aqui estamos gerando um array com uma distribuição normal

normal = sct.norm.rvs(loc=10, scale=1, size=1000)

# aqui estamos gerando um array com uma distribuição t Student

t_student = sct.t.rvs(randomlist, loc=10, scale=1, size=1000)

# aqui estamos gerando um array com uma distribuição qui quadrado

qui_quadrado = sct.chi2.rvs(randomlist, loc=10, scale=1, size=1000)

# aqui estamos gerando um array com uma distribuição uniforme

uniforme = sct.uniform.rvs(loc=0, scale=1, size=1000)
```

```python
sns.distplot(normal)
```

```python
sns.distplot(t_student)
```

```python
sns.distplot(qui_quadrado)
```

```python
sns.distplot(uniforme)
```

```python
type(uniforme)

n=np.random.choice(uniforme,10)

media = np.average(n)

print(media)
```

```python
#Independente da distribuição da população a média das amostras tende a seguir uma normal;

#Uniforme

#Looping criando lista da média das amostras

sample_means=[] #resultado


while len(sample_means) != 50:
    
    n=np.random.choice(uniforme,10) # pega 10 valores aleatórios e guarda
    media = np.average(n)
    sample_means.append(media)
print(sample_means)
```
